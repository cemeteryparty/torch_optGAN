```python
class conf:
    optimizer = "Adam"
    epochs, batch_size, max_iter = 2000, 128, 100
    dis_adv_iter = 1
    saved = "saved"
    g_lr, d_lr = 1e-4, 1e-4
    ckpt=None
    def __init__(self):
        self.d = dict(optimizer=self.optimizer)
        self.d["epochs"] = self.epochs
        self.d["max_iter"] = self.max_iter
        self.d["batch_size"] = self.batch_size
        self.d["dis_adv_iter"] = self.dis_adv_iter
        self.d["saved"] = self.saved
        self.d["g_lr"] = self.g_lr
        self.d["d_lr"] = self.d_lr
        self.d["ckpt"] = self.ckpt
args = conf()

fd = open(os.path.join(args.saved, "conf.json"), "w")
json.dump(args.d, fd, indent=4)
fd.close()
```

```python
def read_json(filepath):
    if filepath is None:
        return None
    fd = open(filepath, "r")
    content = json.load(fd)
    fd.close()
    return content
read_json("saved/conf.json")
```
